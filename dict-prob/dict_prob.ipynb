```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import pandas as pd

#zdf = pd.DataFrame()
SPECZ1=['ㄓ','ㄔ','ㄕ','ㄖ','ㄗ','ㄘ','ㄙ']
SPECZ2=['ㄧ','ㄨ','ㄩ']
CSPECZ2=['i','u','y']
FOURSOUNDS=['ˊ','ˇ','ˋ','˙']
DUOZ=['ㄧㄢ','ㄩㄢ','ㄧㄣ','ㄧㄥ','ㄨㄥ','ㄩㄥ']

def read_zhuyin2ipa_table():
    zdf = pd.read_excel("zhuyin2ipa.xlsx")
    #print(zdf)
    return zdf

def is_ascii(check_str):
     return all(ord(c) < 128 for c in check_str)

def foursounds(c):
    if 'ˋ' in c:
        return "51 "
    elif 'ˊ' in c:
        return "35 "
    elif 'ˇ' in c:
        return "21 "
    elif '˙' in c:
        return "0 "
    else:
        return "55 "
    
def convertZH(zdf, zh):
    '''This function checks the Zhuyin table and return corresponding
       IPA, 
       return '' if not found
       return IPA element in the table with leading and tailing space
              stripped
    '''
    #print("searching", zh)
    x = zdf[zdf['Zhuyin']==zh].IPA.to_string(index=False)
    if x == "Series([], )":
        return ''
    else:
        #return x.replace(' ','')
        return x.rstrip().lstrip().replace('\n','')

def convertZHL(zdf, zh):
    s = convertZH(zdf, zh)
    if s == '':
        return []
    else:
        return s.split(' ')
    
def remove_foursounds(zh):
    z = zh
    for f in FOURSOUNDS:
        z = z.replace(f,'')
    return z

def is_duo(zh):
    for d in DUOZ:
        if d in zh:
            return True
    return False

def duo_convert(zdf, zh):
    #print("zh0",zh[0])
    matching = False
    for z in SPECZ2:
        #print("zzh", z, zh[0])
        if z == zh[0]:
            matching = True
            break
    
    if matching:
        #print("Converting zh", zh)
        return convertZHL(zdf, zh)
    else:
        lead = convertZHL(zdf, zh[0])
        trail = convertZHL(zdf, zh[1:])
        print(lead, trail)
        return lead.append(trail)
    
def convertIUY(z):
    if z == 'ㄧ':
        return 'i'
    elif z == 'ㄨ':
        return 'u'
    else:
        return 'y'
    
def toipa(zdf, zhuyin):
    zh=remove_foursounds(zhuyin)
    idx = 0
    zhlen = len(zh)
    
    if is_duo(zh):
        return(duo_convert(zdf, zh))
    
    IPAL=[]
    while (idx < zhlen):
        #print(idx)
        if zh[idx] in SPECZ1:
            if zhlen == 1:
                return ([convertZHL(zdf, zh[idx]), ['ɨ']])
            else:
                IPAL.append(convertZHL(zdf, zh[idx]))
                idx = idx + 1
                
        elif zh[idx] in SPECZ2:
            if idx == zhlen - 1:
                #The last one
                lc = [convertIUY(zh[idx])]
                IPAL.append(lc)
                return IPAL
            else:
                IPAL.append(convertZHL(zdf, zh[idx]))
                idx = idx + 1
        else:
            IPAL.append(convertZHL(zdf, zh[idx]))
            idx = idx + 1
            
    #print(IPAL)
    return IPAL

```

```python
def read_dict():
    # to read just one sheet to dataframe:
    df = pd.read_excel('cdict_cond.xlsx')
    return(df)
```

```python
def zhuyin2ipa(ddf, zdf):
    for index, row in ddf.iterrows():
        phlist = row['phone'].split('　')
        for p in phlist:
            pipa = toipa(zdf, p)
            print(row['word'], p, pipa, end=" ")
        print()
```
