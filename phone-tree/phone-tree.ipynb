```python
import pandas as pd
import treelib

def read_databases():
    try:
        #讀入字典
        cdict = pd.read_excel('cdict_cond.xlsx')

        return cdict
    except Exception as e:
        print(e)
        exit(-1)

cdict = read_databases()
```

```python
from treelib import Node, Tree

tree = Tree()
root = tree.create_node("root", "root") # root node

for index, row in cdict.iterrows():
    word = row['word']
    phone = row['phone']
    if len(word) == 1:
        #print(p, end=" ")
        tmp=phone
        if "ˊ" in tmp:
            tmp=tmp.replace("ˊ","") +"2"
        elif "˙" in tmp:
            tmp=tmp.replace("˙","") +"0"
        elif "ˇ" in tmp:
            tmp=tmp.replace("ˇ","") +"3"
        elif "ˋ" in tmp:
            tmp=tmp.replace("ˋ","") +"4"
        else:
            tmp=tmp + "1"
        tmp=tmp.replace(" ","")
        tmp=tmp.replace(",","")
        #print(tmp, len(tmp))
        #print('.', end="")
        for j in range(len(tmp)):
            #print(tmp[0:j+1])
            phnode = tree.get_node(tmp[0:j+1])
            #print(phnode)
            if phnode == None:
                if j == 0:
                    px = tree.create_node(tmp[0], tmp[0], parent="root")
                else:
                    phnode = tree.get_node(tmp[0:j])
                    px = tree.create_node(tmp[j], tmp[0:j+1], parent=phnode.identifier)

tree.show()
```
