```python
import tgt
import glob
```

```python
def textgrid_creation(name):
    txtgrname=name.replace('TXT','textgrid')
    txtgr = tgt.TextGrid(txtgrname)    
    #txtgr = txtgr_create(name)
    return(txtgr)
```

```python
def tier_creation(filename, tiername):
    try:
        lines = []
        with open(filename) as f:
             [lines.append(line.rstrip().split(' ')) for line in f.readlines()]

        ntier = tgt.IntervalTier(start_time=float(lines[0][0])/16000., end_time=float(lines[-1][1])/16000., name=tiername)
        for l in lines:
            ann = tgt.Annotation(float(l[0])/16000., float(l[1])/16000., ' '.join(l[2:]))
            ntier.add_annotation(ann)

        return ntier
    except:
        print("Error in file", filename)
        return None
```

```python
TXTF=glob.glob('TEST/*/*/*.TXT')
for name in TXTF:
    print(name)
    txtgr = textgrid_creation(name)
    ttier = tier_creation(name, 'Text')
    wtier = tier_creation(name[:-3]+'WRD', 'Word')
    ptier = tier_creation(name[:-3]+'PHN', 'Phone')
    txtgr.add_tier(ttier)
    if wtier is not None:
        txtgr.add_tier(wtier)
    txtgr.add_tier(ptier)
    tgt.io.write_to_file(txtgr, name[:-3]+'textgrid', format='long', encoding='utf-8')
```
